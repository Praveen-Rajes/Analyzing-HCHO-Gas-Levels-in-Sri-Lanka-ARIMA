```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.patheffects import withStroke

# Define the lockdown period
lockdown_start = '2020-05-01'
lockdown_end = '2021-12-01'

# Extract the HCHO values for each city
bibile_before = Bibile['HCHO Value'][:lockdown_start].mean()*100000
bibile_on = Bibile['HCHO Value'][lockdown_start:lockdown_end].mean()*100000
bibile_after = Bibile['HCHO Value'][lockdown_end:].mean()*100000
colombo_before = Colombo['HCHO Value'][:lockdown_start].mean()*100000
colombo_on = Colombo['HCHO Value'][lockdown_start:lockdown_end].mean()*100000
colombo_after = Colombo['HCHO Value'][lockdown_end:].mean()*100000
deniyaya_before = Deniyaya['HCHO Value'][:lockdown_start].mean()*100000
deniyaya_on = Deniyaya['HCHO Value'][lockdown_start:lockdown_end].mean()*100000
deniyaya_after = Deniyaya['HCHO Value'][lockdown_end:].mean()*100000
jaffna_before = Jaffna['HCHO Value'][:lockdown_start].mean()*100000
jaffna_on = Jaffna['HCHO Value'][lockdown_start:lockdown_end].mean()*100000
jaffna_after = Jaffna['HCHO Value'][lockdown_end:].mean()*100000
kurunegala_before = Kurunegala['HCHO Value'][:lockdown_start].mean()*100000
kurunegala_on = Kurunegala['HCHO Value'][lockdown_start:lockdown_end].mean()*100000
kurunegala_after = Kurunegala['HCHO Value'][lockdown_end:].mean()*100000
nuwara_eliya_before = Nuwara_Eliya['HCHO Value'][:lockdown_start].mean()*100000
nuwara_eliya_on = Nuwara_Eliya['HCHO Value'][lockdown_start:lockdown_end].mean()*100000
nuwara_eliya_after = Nuwara_Eliya['HCHO Value'][lockdown_end:].mean()*100000
kandy_before = Kandy['HCHO Value'][:lockdown_start].mean()*100000
kandy_on = Kandy['HCHO Value'][lockdown_start:lockdown_end].mean()*100000
kandy_after = Kandy['HCHO Value'][lockdown_end:].mean()*100000

# Define the locations and HCHO values
locations = ['Bibile', 'Colombo', 'Deniyaya', 'Jaffna', 'Kurunegala', 'Nuwara Eliya', 'Kandy']
hcho_before = [bibile_before, colombo_before, deniyaya_before, jaffna_before, kurunegala_before, nuwara_eliya_before, kandy_before]
hcho_on = [bibile_on, colombo_on, deniyaya_on, jaffna_on, kurunegala_on, nuwara_eliya_on, kandy_on]
hcho_after = [bibile_after, colombo_after, deniyaya_after, jaffna_after, kurunegala_after, nuwara_eliya_after, kandy_after]

# Define the coordinates for each location
coordinates = {
    'Bibile': (81.3, 7.2),
    'Colombo': (79.8, 6.9),
    'Deniyaya': (80.6, 6.3),
    'Jaffna': (80.1, 9.6),
    'Kurunegala': (80.4, 7.7),
    'Nuwara Eliya': (80.8, 6.9),
    'Kandy': (80.6, 7.3)
}

```

```python


# Create the figure and subplots
fig = plt.figure(figsize=(30, 10), facecolor='white')

# Plot the "Before" values on the first subplot
ax_before = fig.add_subplot(1, 5, 1)
ax_before.set_title('Before Lockdown', fontsize=12)
m = Basemap(llcrnrlon=79.5, llcrnrlat=5.9, urcrnrlon=81.9, urcrnrlat=9.8, resolution='i', projection='merc', ax=ax_before)
m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.fillcontinents(color='white', lake_color='aqua')
for location, coord in coordinates.items():
    x, y = m(coord[0], coord[1])
    txt = plt.text(x, y, f"{location}\nBefore: {hcho_before[locations.index(location)]:.2f}", ha='center', va='center', color='black', fontweight='bold')
    txt.set_path_effects([withStroke(linewidth=2, foreground='white')])

# Plot the "During" values on the second subplot
ax_during = fig.add_subplot(1, 5, 2)
ax_during.set_title('During Lockdown', fontsize=12)
m = Basemap(llcrnrlon=79.5, llcrnrlat=5.9, urcrnrlon=81.9, urcrnrlat=9.8, resolution='i', projection='merc', ax=ax_during)
m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.fillcontinents(color='white', lake_color='aqua')
for location, coord in coordinates.items():
    x, y = m(coord[0], coord[1])
    txt = plt.text(x, y, f"{location}\nDuring: {hcho_on[locations.index(location)]:.2f}", ha='center', va='center', color='black', fontweight='bold')
    txt.set_path_effects([withStroke(linewidth=2, foreground='white')])

# Plot the "After" values on the third subplot
ax_after = fig.add_subplot(1, 5, 3)
ax_after.set_title('After Lockdown', fontsize=12)
m = Basemap(llcrnrlon=79.5, llcrnrlat=5.9, urcrnrlon=81.9, urcrnrlat=9.8, resolution='i', projection='merc', ax=ax_after)
m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.fillcontinents(color='white', lake_color='aqua')
for location, coord in coordinates.items():
    x, y = m(coord[0], coord[1])
    txt = plt.text(x, y, f"{location}\nAfter: {hcho_after[locations.index(location)]:.2f}", ha='center', va='center', color='black', fontweight='bold')
    txt.set_path_effects([withStroke(linewidth=2, foreground='white')])

plt.tight_layout()
plt.show()
```

Population has a positive correlation with every other factor except sea proximity. This means that areas with higher populations tend to have higher levels of sea proximity, formaldehyde (HCHO) and all the yearly mean HCHO values.
Sea proximity has a negative correlation with all yearly mean HCHO values. This means that areas closer to the sea tend to have higer levels of formaldehyde.
Each of the yearly mean HCHO values (2019-2023) has a positive correlation with all the other yearly mean HCHO values. This means that areas with higher formaldehyde levels in one year tend to have higher formaldehyde levels in other years.
